```python
color_assign={
   ' North Indian'      : 'blue',
    'Cafe'              : 'green',
    'Chinese '          : 'red',
   ' Continental'       : 'purple',
    'Italian'           : 'orange',
    'Fast Food'         : 'darkred',
    'Bakery'            : 'lightred',
    'American'          : 'beige',
    'South Indian'      : 'darkblue',
    'Pizza'             : 'darkgreen',
    'Burger'            : 'cadetblue',
    'Mughlai'           : 'darkpurple',
    'European'          : 'pink',
    'Finger Food'       : 'lightblue',
    'Street Food'       : 'lightgreen',
    'Asian'             : 'gray',
    'Modern Indian'     : 'black',
    'Mexican'           : 'darkorange',
    'Desserts'          : 'darkyellow',
    'Seafood'           : 'lightyellow',
    'Mediterranean'     : 'darkturquoise',
    'Biryani'           : 'mediumaquamarine',




}
```

```python
def assign(cuisine):
  if(cuisine in color_assign):
     return color_assign[cuisine]
  else: return 'white'
```

```python
from folium.plugins.semicircle import Marker
c_map=folium.Map([28.582839, 77.201089],titles='cartodbpositron',zoom_start=10)
mc=MarkerCluster().add_to(c_map)
for index,row in coordinates.iterrows():
  folium.Marker(
        location=[row['Latitude'],row['Longitude']],
        popup=row['First_Cuisine'],
        icon=folium.Icon(color=assign(row['First_Cuisine']))).add_to(mc)
```
